```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
#generate a evenly spaced value from zero radians to 2 PI radian
n=200
x=np.linspace(0,2*np.pi,n)
sine_values=np.sin(x)
plt.plot(x,sine_values)
```

```python
##genertae noisy data
noise=0.5
noisy_sine_value=sine_values+np.random.uniform(-noise,noise,n)
plt.plot(x,noisy_sine_value,color="r")
plt.plot(x,sine_values,linewidth=3)
```

```python
##calculate mse using the equation
error_value=(1/n)*sum(np.power(sine_values-noisy_sine_value,2))
error_value
```

```python
#using function from sklearn library
mean_squared_error(sine_values,noisy_sine_value)
```
